```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from random import randint, random
```

```python
class Cell:
    def __init__(self,state):
        """
        parametros:
            state -> el estado, ya sea la altura de la arena, o si el arbol esta vivo, muerto o quemandose
        """
        self.state=state
        
    """
    estadoActual es la vecindad (ya sea de 8 o de 2), ya que de esta depende lo que le suceda al estado de la celda
    """
        
    #todo esto es de arena 1d            
    def nuevoEstadoBorde(self,estadoActual,sc):
        """
        Caso especial para dinámicaCaja. Solo se usa para el borde. No se cae ningun grano de arena, solo acumula los que recibe.
        """
        if estadoActual[0]-self.state>sc: #h_i-1 - h_i
            self.state+=1
        
    
    def nuevoEstadoArena(self,estadoActual,sc):
        """
        Se aplica a todas las columnas excepto a los bordes si estamos simulando una caja
        """
        if self.state-estadoActual[2]>sc: #h_i - h_i+1
            self.state-=1
        if estadoActual[0]-self.state>sc: #h_i-1 - h_i
            self.state+=1

    def agrega(self):
        self.state+=1
    
    def random(self):
        self.state=randint(0,5)
    
    def wipe(self):
        self.state=0
        
    #esto es bosque
    def nuevoEstadoBosque(self, estadoActual, f,p):
        """
        revisa su estado y el estado de los 8 a su alrededor para decidir su proximo estado
        """
        if self.state==1:
            self.checaFuego(estadoActual)
            #if self.state==1:#este if es por si el estado sigue siendo vivo (despues de checar el fuego de los vecinos)
            self.spontaneousCombustion(f)
        else:
            self.muere()
            self.revive(p)
    
    def checaFuego(self, estadoActual):
        """
        revisa que los de alrededor no se esten quemando
        """
        for i in xrange(0,3,1):
            for j in xrange(0,3,1):
                if estadoActual[i,j]==2:
                    self.state=2
    
    def spontaneousCombustion(self,f):
        """
        tiene una probabilidad para arder en llamas repentinamente
        """
        if self.state==1 and random()<=f:
            self.state = 2
    
    def muere(self):
        """
        si el estado es en llamas, ahora esta muerto
        """
        if self.state==2:
            self.state = 0
    
    def revive(self,p):
        """
        hay una probabilidad de que el arbol reviva
        """
        if self.state == 0 and random()<=p: #por que no random.random()??
            self.state=1
            
    def algunasEnLlamas(self):
        if random()<0.30:
            self.state=2
        else:
            self.state=1
            
    #esto es pila de arena2
    def nuevoEstadoArena2(self,estadoActual,K):
        """
        checa no ser mayor a K, o que los 4 vecinos no sean mayores a K
        """
        if self.state>K:
            self.state-=4
        if estadoActual[0,1]>K:
            self.state+=1
        if estadoActual[1,0]>K:
            self.state+=1
        if estadoActual[1,2]>K:
            self.state+=1
        if estadoActual[2,1]>K:
            self.state+=1
    
    """
    las demas funciones de pila de arena 2 son iguales a las de pila de arena 1 y no tenia caso copiarlas dos veces
    """
```

```python
class PilaArena:
    def __init__(self,L):
        """
        parametros:
        L -> longitud del arreglo
        sc -> pendiente critica
        estadoActual -> un arreglo de tamanio L de Cells
        """
        self.L=L
        self.sc=2
        self.estadoActual=np.empty((1,self.L), dtype=object)
        
        self.establesEnTiempo=np.zeros((100,2))
        
    def crea(self):
        """
        llena el arreglo con Cells (cuyo estado es 0)
        """
        for i in xrange(0,self.L,1):
            self.estadoActual[0,i]=Cell(0)
        
    
    def dinamicaMesa(self):
        """
        revisa que la pila sea estable, si lo es, agrega aleatoriamente un grano de arena, 
        si no lo es, cambia el estado de todas las Cells, en base al estado que ya se tiene (para eso "estatico")
        """
        estatico=self.creaEstadoEstatico()
        estable=True
        for i in xrange(1,self.L-1,1):
            if estatico[0,i]-estatico[0,i+1]>self.sc:
                estable=False
        if estable:
            x=randint(1,self.L-1)
            self.estadoActual[0,x].agrega()
        else:
            for i in xrange(1,self.L-1,1):
                self.estadoActual[0,i].nuevoEstadoArena(estatico[0,i-1:i+2],self.sc)
    
    def dinamicaCaja(self):
        """
        revisa que la pila sea estable, si lo es, agrega aleatoriamente un grano de arena, 
        si no lo es, cambia el estado de todas las Cells, en base al estado que ya se tiene (para eso "estatico")
        """
        estatico=self.creaEstadoEstatico()
        estable=True
        for i in xrange(1,self.L-2,1):
            if estatico[0,i]-estatico[0,i+1]>self.sc:
                estable=False
        if estable:
            x=randint(1,self.L-1)
            self.estadoActual[0,x].agrega()
        else:
            for i in xrange(1,self.L-2,1):
                self.estadoActual[0,i].nuevoEstadoArena(estatico[0,i-1:i+2],self.sc)
            self.estadoActual[0,self.L-2].nuevoEstadoBorde(estatico[0,self.L-3:self.L],self.sc)
            """
            A diferencia de dinamicaMesa, este proceso separa el borde derecho y lo manda a nuevoEstadoBorde para que acumule los
            granos de arena que recibe.
            """
   
    def creaEstadoEstatico(self):
        """
        crea un arreglo con los estados de las Cells (no las Cells en si, ya que solo son necesarios los nombres) 
        """
        estatico=np.empty((1,self.L))
        i=0
        for celda in self.estadoActual[0,:]:
            estatico[0,i]=celda.state
            i+=1
        return estatico
    
    def toString(self):
        """
        crea un arreglo con los estados de las Cells para imprimirlo
        ya que imprimir un arreglo de objetos te da sus direcciones en memoria
        """
        imprime=np.empty((1,self.L))
        for i in xrange(0,self.L,1):
            imprime[0,i]=self.estadoActual[0,i].state
        print imprime[0,1:self.L-1]
        #print self.estadoActual
    
    def startRandom(self):
        """
        se asigna valor aleatorio a las Cells 
        """
        for i in xrange(1,self.L-1,1):
            self.estadoActual[0,i].random()
            
    def wipe(self):
        """
        borra las orillas (que de hecho nunca se despliegan)
        esto permite que la pila de arena se desparrame como si estuviera sobre una mesa
        """
        self.estadoActual[0,0].wipe()
        self.estadoActual[0,self.L-1].wipe()
        
    def loopMesa(self, steps):
        """
        Ejecuta el numero especificado de pasos
        """
        for i in xrange(0,steps):
            self.cuentaEstables(i)
            self.stepMesa()
    
    def loopCaja(self, steps):
        """
        Ejecuta el numero especificado de pasos
        """
        #[self.stepCaja() for i in xrange(steps)]
        for i in xrange(0,steps):
            self.cuentaEstables(i)
            self.stepCaja()
        

    def stepMesa(self):
        """
        Avanza un paso t -> t+1
        """
        self.dinamicaMesa()
        self.wipe()
    
    def stepCaja(self):
        """
        Avanza un paso t -> t+1
        """
        self.dinamicaCaja()
        self.wipe()
        
    def cuentaEstables(self,t):
        estatico=self.creaEstadoEstatico()
        inestables=0
        for i in xrange(1,self.L-2,1):
            if estatico[0,i]-estatico[0,i+1]>self.sc:#no estable
                inestables+=1
        self.establesEnTiempo[t,0]=t
        self.establesEnTiempo[t,1]=inestables
        return self.establesEnTiempo
```

```python
class PilaArena2:
    def __init__(self,L):
        """
        parametros:
            L -> longitud de los lados del arreglo (el arreglo tiene LxL celdas)
            estadoActual -> un arreglo de tamanio L vacio de tipo object para llenarlo de Cells
            K -> estado critico
        """
        self.L=L
        self.estadoActual=np.empty((self.L,self.L), dtype=object)
        self.K=4
        self.inestablesEnTiempo=np.zeros((100,2))
    
    def crea(self):
        """
        llena el arreglo con Cells (cuyo estado es 0)
        """
        for i in xrange(0,self.L,1):
            for j in xrange(0,self.L,1):
                self.estadoActual[i,j]=Cell(0)
    
    def creaEstadoEstatico(self):
        estatico=np.empty((self.L,self.L))
        for i in xrange(0,self.L,1):
            for j in xrange(0,self.L,1):
                estatico[i,j]=self.estadoActual[i,j].state
        return estatico
    
    def dinamica(self):
        """
        revisa que la pila sea estable, si lo es, agrega aleatoriamente un grano de arena, 
        si no lo es, cambia el estado de todas las Cells, en base al estado que ya se tiene (para eso "estatico")
        """
        estatico=self.creaEstadoEstatico()
        estable=True
        for i in xrange(1,self.L-1,1):
            for j in xrange(1,self.L-1,1):
                if estatico[i,j]>self.K:
                    estable=False
                    #print "falso", estatico[i,j]
        if estable:
            self.estadoActual[randint(1,self.L-2),randint(1,self.L-2)].agrega()
        else:
            #print "inestable"
            for i in xrange(1,self.L-1,1):
                for j in xrange(1,self.L-1,1):
                    pedazo=estatico[i-1:i+2,j-1:j+2]
                    self.estadoActual[i,j].nuevoEstadoArena2(pedazo,self.K)
    
    def toString(self):
        """
        crea un arreglo con los estados de las Cells para imprimirlo
        ya que imprimir un arreglo de objetos te da sus direcciones en memoria
        """
        imprime=np.empty((self.L,self.L))
        for i in xrange(0,self.L,1):
            for j in xrange(0,self.L,1):
                imprime[i,j]=self.estadoActual[i,j].state
        print imprime[1:self.L-1,1:self.L-1]
        
    def startRandom(self):
        """
        se asigna valor aleatorio a las Cells 
        """
        for i in xrange(1,self.L-1,1):
            for j in xrange(1,self.L-1,1):
                self.estadoActual[i,j].random()
    
    def wipe(self):
        """
        borra las orillas (el marco)
        esto permite que la pila de arena se desparrame como si estuviera sobre una mesa
        """
        for i in xrange(0,self.L-1):
            self.estadoActual[i,0].wipe()
            self.estadoActual[0,i].wipe()
            self.estadoActual[self.L-1,i].wipe()
            self.estadoActual[i,self.L-1].wipe()
        """self.estadoActual[:,0].wipe()
        self.estadoActual[0,:].wipe()
        self.estadoActual[L-1,:].wipe()
        self.estadoActual[:,L-1].wipe()"""
    
    def loop(self, steps):
        """
        Ejecuta el número especificado de pasos
        """
        for i in xrange(0,steps):
            self.cuentaEstables(i)
            self.step()

    def step(self):
        """
        Avanza un paso t -> t+1
        """
        self.dinamica()
        self.wipe()
    
    def cuentaEstables(self,t):
        estatico=self.creaEstadoEstatico()
        inestables=0
        for i in xrange(1,self.L-1,1):
            for j in xrange(1,self.L-1,1):
                if estatico[i,j]>self.K:
                    inestables+=1
        self.inestablesEnTiempo[t,0]=t
        self.inestablesEnTiempo[t,1]=inestables
        return self.inestablesEnTiempo
```
